```python
# 255  gray
img = mpimg.imread('Baboon.jpg')
R, G, B = img[:, :, 0], img[:, :, 1], img[:, :, 2]
gray_img_transf_1 = 0.2989 * R + 0.587 * G + 0.114 * B

transf_1 = 255 - gray_img_transf_1
image_1 = np.array(transf_1, dtype=np.uint8)
plt.imshow(image_1)
plt.show()
plt.imshow(gray_img_transf_1, cmap='gray')
plt.show()
```

```python
# gray + 100
img = mpimg.imread('Baboon.jpg')
R, G, B = img[:, :, 0], img[:, :, 1], img[:, :, 2]
gray_img_transf_2 = 0.2989 * R + 0.587 * G + 0.114 * B

transf_2 = gray_img_transf_2 + 100
image_2 = np.array(transf_2, dtype=np.uint8)
plt.imshow(image_2)
plt.show()
plt.imshow(gray_img_transf_2, cmap='gray')
plt.show()

```

```python
# sqtr(gray,1/2)
img = mpimg.imread('Baboon.jpg')
R, G, B = img[:, :, 0], img[:, :, 1], img[:, :, 2]
gray_img_transf_3 = 0.2989 * R + 0.587 * G + 0.114 * B

transf_3 = np.power(gray_img_transf_3,(1/2))
image_3 = np.array(transf_3, dtype=np.uint8)
plt.imshow(image_3)
plt.show()
plt.imshow(gray_img_transf_3, cmap='gray')
plt.show()
```
